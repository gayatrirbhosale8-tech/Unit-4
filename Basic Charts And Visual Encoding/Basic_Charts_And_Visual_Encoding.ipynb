```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {
    'Days':['Monday','Tuesday','Wednesday','Thursday','Friday'],
    'Study_Hours': [2,3,2.5,4,3.5],
    'Marks':[50,60,55,70,65],
    'Attendance':[60,70,65,80,75],
    'Sleep Hours':[6,7,6,8,7],
    'Assignments_completed':[1,2,1,3,2]
}

df = pd.DataFrame(data)
print(df)
```

```python
#1. Line Chart (Trend)

plt.plot(df['Days'],df['Study_Hours'],marker='o')
plt.show()
```

```python
plt.plot(df['Days'],df['Study_Hours'],marker='o')
plt.title("Study Hours Over Days")
plt.xlabel("Days")
plt.ylabel("Study Hours")
plt.show()
```

```python
#Avance Line Chart

plt.figure(figsize=(7,4))
plt.plot(df['Days'],df['Study_Hours'],
         marker='o',color = 'LightGreen',label = 'Study_Hours')
plt.plot(df["Days"],df["Marks"],
         marker='o',color = 'Orange',label = 'Marks')
plt.title("Study Hours and Marks Over Days")
plt.xlabel("Days")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
#2 Bar Chart(Comparison)
plt.bar(df['Days'],df['Marks'],color='Sky Blue')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.title('Marks comparison')
plt.show()
```

```python
# Advance Bar Graph
bars = plt.bar(df['Days'],df['Marks'],color='lavender')
for bar in bars:
  y = bar.get_height()
  plt.text(bar.get_x() + bar.get_width()/2,y,str(y),
           ha='center',va='bottom')


plt.xlabel('Days')
plt.ylabel('Marks')
plt.title('Marks comparison')
plt.grid(axis='y')    #adds horizontal grid for readability
plt.show()
```

```python
#Histogram
#Show how values are spread

data = {
    'Marks':[50,60,55,70,65,75,80,85,90,45,68,72,77,82,88]
}
df =  pd.DataFrame(data)
plt.hist(df['Marks'],bins=7,edgecolor='black',color='pink')
plt.xlabel
plt.title('Marks Distribution')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
from matplotlib import colors
import pandas as pd

# Re-define the original data and DataFrame to ensure 'Study_Hours' and 'Marks' are available
data_for_plot = {
    'Days':['Monday','Tuesday','Wednesday','Thursday','Friday'],
    'Study_Hours': [2,3,2.5,4,3.5],
    'Marks':[50,60,55,70,65],
    'Attendance':[60,70,65,80,75],
    'Sleep Hours':[6,7,6,8,7],
    'Assignments_completed':[1,2,1,3,2]
}
df = pd.DataFrame(data_for_plot)

#Conditional scatter plot
#Example category
df['Result']=['Fail','Pass','Fail','Pass','Pass']
colors = ['red' if result == 'Fail' else 'green' for result in df['Result']]
plt.scatter(df['Study_Hours'],df['Marks'],c=colors)
plt.xlabel('Study_Hours')
plt.ylabel('Marks')
plt.title('Study_Hours vs Marks')
```
