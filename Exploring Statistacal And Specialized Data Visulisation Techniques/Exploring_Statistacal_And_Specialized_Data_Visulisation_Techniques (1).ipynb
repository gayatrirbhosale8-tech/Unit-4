```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)
df = pd.DataFrame({
    'Category':['A','B','C','D','E'],
    'Values':[50,70,30,90,60],
    'Sales': np.random.randint(100,500,5),
    'Profit': np.random.randint(10,100,5)
})
print(df.head())
```

```python
import matplotlib.pyplot as plt
#AreaPlot
plt.figure()
plt.fill_between(df['Category'],df['Values'],color='Grey',alpha = 1)
plt.title('Area Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

```python
# Pie Chart
plt.figure()
plt.pie(df['Values'],labels=df['Category'],autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
# Donut Chart
plt.figure()
plt.pie(df['Values'],labels=df['Category'],autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})
centre_circle = plt.Circle((0,0),0.20,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.show()
```

```python
#4 Boxplot (outliners detection)
plt.figure()
plt.boxplot(df['Values'])
plt.title('Boxplot of Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure()
plt.boxplot(x=df['Values'])
plt.title('Boxplot of Sales')
plt.show()
```

```python
#5 HeatMap (Correaltion)
plt.figure()
corr= df[['Sales','Profit']].corr()
sns.heatmap(corr,annot=True,cmap='inferno')
plt.title("Correlation Heatmap")
plt
```

```python
#Bubble plot
import matplotlib.pyplot as plt
plt.scatter(df['Sales'],df['Profit'],
            s=df['Values']*10,c='purple',alpha = 0.5)
plt.title('Bubble Plot')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.scatterplot(x='Sales',y='Profit',
                size='Values',hue='Values',
                sizes=(50,300),palette='spring',
                data=df)
plt.title('Bubble Plot')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```
